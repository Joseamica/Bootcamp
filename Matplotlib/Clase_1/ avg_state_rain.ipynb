```python
# Dependencies
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import numpy as np
import pandas as pd
```

```python
# Load in csv
df = pd.read_csv('avg_rain_state.csv')
df.head()
```

```python
# Set x axis and tick locations
plt.figure(figsize=(20,1))
x_axis = df['State']
y_axis = df['Inches']
plt.yticks(np.arange(0,58,25))
plt.bar(x_axis,y_axis, color='r', alpha=0.5, align="edge")
c = [value+0.4 for value in range(len(x_axis))]
plt.xticks(c , x_axis, rotation='vertical')
plt.title('Average Rain per State')
plt.xlabel('State')
plt.ylabel('Average Amount of Rainfall in lr')
```
