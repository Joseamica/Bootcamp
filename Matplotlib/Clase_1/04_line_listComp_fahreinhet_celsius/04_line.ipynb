```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
```

```python
fah = [39, 42, 51, 62, 72, 82, 86, 84, 77, 65, 55, 44]
celsius = []  
    
celsius = [(i-32)*0.56 for i in fah]

#hace una lista del 1 al 12 de 1 a 1.
x_axis = np.arange(1,13,1)
```

```python
plt.plot(x_axis, fah)
plt.plot(x_axis, celsius)
plt.xlabel('Month')
plt.ylabel('Temperature')
```

```python
fah_handle = plt.plot(x_axis, fah, marker = 'o', color='blue', label='Fahreniheit')
celsius_handle = plt.plot(x_axis, celsius, marker = 'o', color='red', label='Celsius')
```

```python
plt.legend(loc="lower right")
```
