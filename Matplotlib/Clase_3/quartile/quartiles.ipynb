```python
# Dependencies
import matplotlib.pyplot as plt
from stats import median
import numpy as np
```

```python
### Data Points
arr = np.array([2.3, 10.2,11.2, 12.3, 14.5, 14.6, 15.0, 15.1, 19.0, 24.0])
arr
```

```python
# Find the median
arr_median = np.median(arr)
arr_median
```

```python
# Use numpy to create quartiles
lower_quartile = np.percentile(arr,25)
high_quartile = np.percentile(arr,75)

iq = high_quartile-lower_quartile
iq
```

```python
# Print the quartiles
```

```python
# Calculate the interquartile range
iq
```

```python
# Find lower boundary
# Q1 - 1.5 * IQR
lower_boundary = lower_quartile - 1.5 * iq
lower_boundary
```

```python
# Find upper boundary
# Q3 + 1.5 * IQR

upper_boundary = high_quartile + 1.5 * iq
upper_boundary
```

```python
# Check for any lower outliers
np.sort(arr)

```

```python
# Check for any upper outliers
np.sort(arr)
upper_outliers = arr[-1]
```

```python
# Create box plot
plt.boxplot(arr, autorange=True)
```
