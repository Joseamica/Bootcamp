```python
# Import Dependencies
import pandas as pd
import numpy as np
```

```python
# Create reference to CSV file
csv_path = "soccer.csv"

# Import the CSV into a pandas DataFrame
soccer_df = pd.read_csv(csv_path, low_memory=False)
soccer_df
```

```python
# Collect a list of all the unique values in "Preferred Position"
soccer_df['Preferred Position'].unique()
```

```python
# Looking only at strikers (ST) to start
soccer_df.sort_values(['ST','Overall'], ascending=False)
```

```python
# Sort the DataFrame by the values in the "ST" column to find the worst
st_worst_df = soccer_df.sort_values(['ST'], ascending=True)
st_worst_df.head(10)
# Reset the index so that the index is now based on the sorting locations
st_worst_df.reset_index(inplace=True)
```

```python
# Save all of the information collected on the worst striker
st_worst_df.to_excel("worst_st.xlsx", index=False)
```
