## Connecting to DB
First we imported sqlalchemy which imports the engines to read sqls
Then we create a file databaseconfig.py with the hidden information

```python
import sqlalchemy as db
import databaseconfig as cfg
user = cfg.mysql['user']
passwd = cfg.mysql['passwd']
engine = db.create_engine(f'postgresql://{user}:{passwd}@localhost:5432/homework_db')
connection = engine.connect()
```

## Importing libraries and creating dataframe
**Step 1)** Read the sql using a code selecting the avg salaries per title
**Step 2)** Create a dataframe using pandas with the previous information
**Step 3)** Plot the dataframe

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python

df = pd.read_sql('select titles.title, AVG(salaries.salary) from salaries join titles on salaries.emp_no = titles.emp_no group by titles.title order by avg(salaries.salary) DESC;', engine)
df = pd.DataFrame(df)

plt.figure(figsize=(15,5))

plt.bar(df.title,df.avg)
plt.title('Average Salaries Per Title')
plt.xlabel('Titles')
plt.ylabel('Salaries')
```
