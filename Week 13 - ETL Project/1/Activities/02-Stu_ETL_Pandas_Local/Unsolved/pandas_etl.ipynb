```python
import pandas as pd
from sqlalchemy import create_engine
```

### Extract CSVs into DataFrames

```python
premise_file = "../Resources/LicensePremise.csv"
```

```python
county_file = "../Resources/CountyLicenseCount.csv"
```

### Transform premise DataFrame

```python
premise_df = pd.read_csv(premise_file)
```


### Transform county DataFrame

```python
county_df = pd.read_csv(county_file)
```

### Create database connection

```python
db_connection = "postgres:aloha1993@localhost:5432/county"
engine = create_engine(f'postgresql://{db_connection}')
```

```python
# Confirm tables
engine.table_names()
```

### Load DataFrames into database

```python
county_df = county_df.rename(columns={"ID": "id", "County Name (Licensee)": "county_name", "County ID Code":"county_id_code", "License Count": "license_count"})
```

```python
premise_df = premise_df.rename(columns={"County ID Code":"county_df","License Received":"license_received","License Certificate Number":"certificate_number","Premises Name":"premises_name","License Serial Number":"serial_number","License Received Date": "License Received"})
premise_df
```

```python
county_df.to_sql(name='county', con=engine, if_exists='append', index=False)
```

```python
premise_df.to_sql(name='premise', con=engine, if_exists='append', index=False)


```
