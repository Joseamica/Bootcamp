```python
import pandas as pd
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
```

```python
# The path to our CSV file
df = pd.read_csv('kickstarter.csv')
# Read our Kickstarter data into pandas
df.head()
```

```python
# Get a list of all of our columns for easy reference
list_df = list(df.columns)
list_df
```

```python
# Extract "name", "goal", "pledged", "state", "country", "staff_pick",
# "backers_count", and "spotlight"
new_df = df[['name', 'goal','pledged','state', 'country', 'staff_pick', 'backers_count', 'spotlight']]
new_df
```

```python
# Remove projects that made no money at all
profitable_df = new_df.loc[new_df['pledged'] > 0 ]
profitable_df
```

```python
# Collect only those projects that were hosted in the US
us_projects = profitable_df.groupby(new_df['country'])
#us_projects.apply(list)
# Create a list of the columns

# Create a new df for "US" with the columns above. 
us_profitable = profitable_df.loc[profitable_df['country'] == "US"]
us_profitable
```

```python
# Create a new column that finds the average amount pledged to a project
us_profitable["Average Pledged"] = us_profitable.loc[:,'pledged'] / us_profitable.loc[:,'backers_count']
us_profitable
```

```python
# First convert "average_donation", "goal", and "pledged" columns to float

# Then Format to go to two decimal places, include a dollar sign, and use comma notation
us_profitable["goal"] = us_profitable["goal"].map("${:.2f}".format)
us_profitable["Average Pledged"] = us_profitable["Average Pledged"].map("${:2f}".format)

us_profitable
```

```python
# Calculate the total number of backers for all US projects
total_us_backers = us_profitable['backers_count'].sum()
total_us_backers
```

```python
# Calculate the average number of backers for all US projects
avg_us_backers = us_profitable['backers_count'].mean()
avg_us_backers
```

```python
# Collect only those US campaigns that have been picked as a "Staff Pick"
us_staff_pick_true = us_profitable.loc[us_profitable['staff_pick'] == True]
us_staff_pick_true
```
