```python
# Create a bunch of samples, each with div items
div = 20
lim = len(housing_data) // div
samples = [housing_data.iloc[(i * div):(i * div + div), 13]
           for i in range(0, lim)]
housing_data = housing_data.rename(columns={0:'CRIM', 1:'ZN',2:'INDUS',3:'CHAS',4:'NOX', 5:'RM', 6:'AGE',7:'DIS',8:'RAD', 9:'TAX',
                            10:'PTRATIO',11:'B',12:'LSTAT',13:'MEDV'})
```
